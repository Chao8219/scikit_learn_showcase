# About

This notebook is to save data into `sklearn.utils.Bunch`.

```python
import numpy as np
from numpy.random import default_rng
import pandas as pd
from sklearn.utils import Bunch
from matplotlib import pyplot as plt
import os

# random number generation
rng = default_rng(23)
```

```python
# find where the data is stored
repo_path = os.path.dirname(os.getcwd())
data_path = repo_path + '\\data\\create_Bunch\\'
```

# Create Data

To test the `Bunch`, we need to create the data first.

```python
bunch_obj = Bunch()
bunch_obj.DESCR = \
'This data is a test dataset to check out the functionality of sklearn.utils.Bunch\n' + \
'More details can be seen in the jupyter notebook.'
bunch_obj.x_arr = rng.uniform(-5.0, 5.0, (101, 1))
bunch_obj.y_arr = np.sin(bunch_obj.x_arr) + rng.normal(0.0, 0.5, (101, 1))

plt.scatter(
    bunch_obj.x_arr, bunch_obj.y_arr, s=5, 
    marker='o', fc='none', ec='blue'
)
plt.tight_layout()
```

# Save and Load Data

```python
np.savez(data_path+'test.npz', bunch_obj=bunch_obj)
with np.load(data_path+'test.npz', allow_pickle=True) as file_loaded:
    bunch_loaded = file_loaded['bunch_obj'][()]
print(bunch_loaded.keys())
print(bunch_loaded.DESCR)
plt.scatter(
    bunch_loaded.x_arr, bunch_loaded.y_arr, s=5, 
    marker='o', fc='none', ec='green'
)
plt.tight_layout()
```
