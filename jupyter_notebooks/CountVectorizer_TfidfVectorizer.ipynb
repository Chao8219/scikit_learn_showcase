This notebook is to check the difference between Count Vectorizer and TF-IDF Vectorizer from sklearn for NLP.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
```

```python
sample_words = [
    'MaryLou wore the tiara with pride', 
    'They both knew it, but neither has the courage or strength to address the underlying issue.', 
    'One dollar and eighty-seven cents.', 
    'Her fingers still touched nothing but air.', 
    'You want to look, but you don\'t at the same time.'
]
```

Count Vectorizer

```python
count_vectorizer = CountVectorizer()
x = count_vectorizer.fit_transform(sample_words)
count_vectorizer_df = pd.DataFrame(x.toarray(), columns=count_vectorizer.get_feature_names_out())
```

```python
count_vectorizer_df
```

TF-IDF Vectorizer

```python
tf_idf_vectorizer = TfidfVectorizer()
x = tf_idf_vectorizer.fit_transform(sample_words)
tf_idf_vectorizer_df = pd.DataFrame(x.toarray(), columns=tf_idf_vectorizer.get_feature_names_out())
```

```python
tf_idf_vectorizer_df
```
